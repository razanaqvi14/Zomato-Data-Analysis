```python
sns.heatmap(df.isnull(), yticklabels = False)
plt.show()
```

```python
df_country = pd.read_excel("Data/Country-Code.xlsx")
df_country
```

```python
df = pd.merge(df, df_country, on = "Country Code", how = "left")
df.head(3)
```

```python
df.dtypes
```

```python
df.columns
```

```python
countries = df["Country"].value_counts().index
# .index makes a list of all the the unique values present inside a column
```

```python
countries_count = df["Country"].value_counts().values
# .values makes a list of how many times a unique value appeared in that column 
```

```python
# Making a pie chart of top 3 Countries where zomato is used the most
plt.pie(countries_count[:3], labels = countries[:3], autopct='%.0f%%')
plt.title("Top 3 Countries Where Zomato is used the most")
plt.show()
```

Observation from the above pie chart
1. The top 3 countries are India, US, UK
2. India covers 94% of zomato, US 5%, and UK 1%

```python
rating_info = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: "No Of Ratings"})
# This is a DataFrameGroupBy, and for here to see that which value occurs how many times you have to use .size() and to convert this into a datafram just add .reset_index()
rating_info
```

Observations from the above table
1. The rating 0 lies in Not Rated
2. The ratings from 1.8 - 2.4 lies in Poor
4. The ratings from 2.5 - 3.4 lies in Average
3. The ratings from 3.5 - 3.9 lies in Good
5. The ratings from 4.0 - 4.4 lies in Very Good
6. The ratings from 4.5 - 4.9 lies in Excellent 

```python
sns.barplot(data = rating_info, x = "Aggregate rating", y = "No Of Ratings", hue = "Rating color", palette=["blue", "red", "orange", "yellow", "green", "green"])
plt.title("Ratings")
# plt.xticks(rotation = 45)
plt.show()
```

## Observations from the above plot
1. The number of people which didn't rate is 2000 +
2. Among people who rated, most people voted between 2.5 - 3.4
3. Most rating the people gave is 3.2 

```python
# Count plot
sns.countplot(data = rating_info, x = "Rating color", palette = ["White", "red", "orange", "yellow", "green", "green"], edgecolor = "black")
plt.title("Ratings According to the dataa")
plt.ylabel("Number Of Rows in data")
plt.show()
```

```python
# Find the country name that has given 0 rating
```

```python
zeroRatingCountries = df[df["Aggregate rating"] == 0].groupby("Country").size().reset_index()
sns.barplot(data = zeroRatingCountries, x = "Country", y = 0)
plt.show()
```
